## Cleaning Data - dropping columns

```python
columns_to_drop = ['id','host_name','last_review']
listing_df.drop(columns_to_drop, axis='columns', inplace=True)
listing_df.isnull().sum()
```

## Cleaning data - replacing NaNs

```python
listing_df.fillna({'reviews_per_month': 0}, inplace=True)
listing_df.isnull().sum()
```

## Visualisations with seaborn

```python
#
sns.countplot(data=listing_df, x='neighbourhood_group')
```

```python
ordered = listing_df['neighbourhood_group'].value_counts().index
sns.countplot(data=listing_df, x='neighbourhood_group', order=ordered);
```

## What is the distribution of flat prices?

```python
sns.distplot(listing_df['price']);
```

```python
# creating a condition for our data to remove outliers
listing_df['price'] <= 500
```

```python
affordable_df = listing_df[listing_df['price'] <=500]
affordable_df
```

```python
sns.distplot(affordable_df['price']);
```

## What is the distribution of flat prices based on the neighbourhood group?

```python
sns.violinplot(data=affordable_df, x='neighbourhood_group', y='price');
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,8))
sns.violinplot(data=affordable_df, x='neighbourhood_group', y='price');
```

# Plotting the listing on a map

```python
affordable_df.plot(
    kind='scatter',
    x = 'longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha=0.8,
    figsize=(12,8));
```

## Adding the NYC map underneath the scatter diagram

```python
import urllib
```

```python
i = urllib.request.urlopen('https://upload.wikimedia.org/wikipedia/commons/e/ec/Neighbourhoods_New_York_City_Map.PNG')
plt.imshow(plt.imread(i), zorder=0, extent=[-74.258, -73.7, 40.49, 40.92])
ax = plt.gcs()

affordable_df.plot(
    kind='scatter',
    x = 'longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha=0.8,
    figsize=(12,8));
```
